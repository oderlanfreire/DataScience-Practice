# Valor de empenho e PIB de municípios gaúchos

```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dados/dados.csv', sep=';')
df.head()
```

```python
srn.histplot(df.PIB, kde=True, bins=10).set(title='PIB')
```

```python
grupo = df.sort_values('PIB').head(15)
grupo = grupo.iloc[:,1:3]
grupo
grupo.plot.bar(x="MUNICIPIO", y="PIB", color="gray")
```

```python
srn.histplot(df.VALOREMPENHO, kde= True, bins=10).set(title="VALOREMPENHO")
```

```python
grupo = df.sort_values('VALOREMPENHO').head(15)
grupo = grupo.iloc[:,[1,3]]
grupo
grupo.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'gray')
```
